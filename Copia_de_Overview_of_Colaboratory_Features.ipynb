```
import random
from fractions import Fraction
import math

# Definimos una clase para representar magnitudes
class Magnitud:
    def __init__(self, nombre, valor, tipo):
        self.nombre = nombre
        self.valor = valor
        self.tipo = tipo  # Puede ser 'positiva' o 'negativa'

    def __str__(self):
        return f"{self.nombre} ({self.tipo}): {self.valor}"

# Creamos una lista de magnitudes ajustadas a la población
cesta_mundo = [
    Magnitud("Salud", 7, 'positiva'),
    Magnitud("Catástrofes", 8, 'negativa'),
    Magnitud("Pandemias", 6, 'negativa'),
    Magnitud("Accidentes", -2.5, 'negativa'),
    Magnitud("Muertes de Hombres", 5, 'positiva'),
    Magnitud("Muertes de Mujeres", 10, 'negativa')  # Suponemos más muertes de mujeres
]

# Función para mostrar las magnitudes
def mostrar_magnitudes(magnitudes):
    for magnitud in magnitudes:
        print(magnitud)

# Función para realizar la sustracción aritmética con manejo de casos especiales
def sustraccion_aritmetica(minuendo, sustraendo):
    resultado = minuendo.valor - sustraendo.valor
    if resultado < 0:
        parte_restante = -resultado
        print(f"Operación imposible: sustraendo {sustraendo.valor} es mayor que minuendo {minuendo.valor}.")
        print(f"Restamos la parte igual ({minuendo.valor}) y queda un número negativo ({-parte_restante}).")
        print(f"El número negativo {-parte_restante} cumple la relación fundamental porque {minuendo.valor} - {sustraendo.valor} = {resultado}.")
    return resultado

# Ejemplo de uso de la sustracción aritmética
minuendo = Magnitud("Salud", 7, 'positiva')
sustraendo = Magnitud("Catástrofes", 8, 'negativa')
resultado_sustraccion = sustraccion_aritmetica(minuendo, sustraendo)
print(f"\nResultado de la Sustracción: {resultado_sustraccion}")

# Función para asignar valores continuos a las magnitudes
def asignar_valores_continuos(magnitudes):
    for magnitud in magnitudes:
        magnitud.valor = round(random.uniform(0.1, 10.0), 2)

# Asignamos valores continuos a las magnitudes y mostramos el resultado
asignar_valores_continuos(cesta_mundo)
print("\nMagnitudes con Valores Continuos:")
mostrar_magnitudes(cesta_mundo)

# Función para medir el conjunto continuo y obtener un número conmensurable
def medir_conjunto(magnitudes):
    total_valor = sum(magnitud.valor for magnitud in magnitudes)
    return round(total_valor / len(magnitudes), 2)

# Contamos unidades discretas y mostramos el resultado
def contar_unidades_discretas(magnitudes, unidad_objeto):
    conteo = sum(magnitud.valor // unidad_objeto for magnitud in magnitudes)
    return conteo

# Contamos unidades continuas y mostramos el resultado
def contar_unidades_continuas(magnitudes, unidad_medida):
    conteo = sum(magnitud.valor / unidad_medida for magnitud in magnitudes)
    return round(conteo, 2)

# Convertimos las mediciones a fracciones
def medir_fracciones(magnitudes, unidad_medida):
    fracciones = [Fraction(magnitud.valor).limit_denominator() for magnitud in magnitudes]
    total_fraccion = sum(fracciones)
    return total_fraccion / Fraction(unidad_medida)

# Medimos con números irracionales
def medir_irracionales(magnitudes, unidad_medida):
    irracionales = [math.sqrt(magnitud.valor) for magnitud in magnitudes]
    total_irracional = sum(irracionales)
    return round(total_irracional / unidad_medida, 2)

# Medimos con números racionales
def medir_racionales(magnitudes, unidad_medida):
    racionales = [Fraction(magnitud.valor).limit_denominator() for magnitud in magnitudes]
    total_racional = sum(racionales)
    return total_racional / Fraction(unidad_medida)

# Manejo de casos especiales
def manejar_casos_especiales(magnitudes, unidad_medida):
    for magnitud in magnitudes:
        if magnitud.valor % unidad_medida != 0:
            print(f"Advertencia: {magnitud.nombre} no contiene exactamente la unidad o sus partes.")
            magnitud.valor = round(magnitud.valor / unidad_medida) * unidad_medida
            print(f"Solución propuesta: Ajustar el valor a {magnitud.valor}.")

# Mostrar el funcionamiento del número negativo en la aritmética
def ejemplo_numero_negativo(minuendo, sustraendo):
    resultado = sustraccion_aritmetica(minuendo, sustraendo)
    if resultado < 0:
        print(f"\nEl número negativo {resultado} cumple la propiedad fundamental de la operación de resta.")
        print(f"Esto demuestra que a pesar de la imposibilidad aritmética, el número negativo representa la parte restante del sustraendo.")
    else:
        print(f"\nResultado de la resta: {resultado}")

# Ejemplo del uso del número negativo
minuendo_negativo = Magnitud("Muertes de Hombres", 5, 'positiva')
sustraendo_negativo = Magnitud("Muertes de Mujeres", 10, 'negativa')
ejemplo_numero_negativo(minuendo_negativo, sustraendo_negativo)

# Comparación de muertes de mujeres y hombres para obtener un resultado negativo
resultado_comparacion = sustraccion_aritmetica(minuendo_negativo, sustraendo_negativo)
print(f"\nResultado de la Comparación de Muertes (Hombres - Mujeres): {resultado_comparacion}")

# Medimos el conjunto continuo y mostramos el resultado
resultado_medicion = medir_conjunto(cesta_mundo)
print(f"\nValor Conmensurable del Conjunto: {resultado_medicion}")

# Número de Unidades Discretas
unidad_objeto = 2  # Unidad arbitraria para las cantidades discretas
resultado_conteo_discreto = contar_unidades_discretas(cesta_mundo, unidad_objeto)
print(f"\nNúmero de Unidades Discretas: {resultado_conteo_discreto}")

# Número de Unidades Continuas
unidad_medida = 1.0  # Unidad arbitraria para las cantidades continuas
resultado_conteo_continuo = contar_unidades_continuas(cesta_mundo, unidad_medida)
print(f"Número de Unidades Continuas: {resultado_conteo_continuo}")

# Medición en Fracciones
resultado_fracciones = medir_fracciones(cesta_mundo, unidad_medida)
print(f"\nMedición en Fracciones: {resultado_fracciones}")

# Medición con Números Irracionales
resultado_irracionales = medir_irracionales(cesta_mundo, unidad_medida)
print(f"Medición con Números Irracionales: {resultado_irracionales}")

# Medición con Números Racionales
resultado_racionales = medir_racionales(cesta_mundo, unidad_medida)
print(f"\nMedición con Números Racionales: {resultado_racionales}")

# Manejo de casos especiales y solución
print("\nManejo de Casos Especiales:")
manejar_casos_especiales(cesta_mundo, unidad_medida)

# Mostramos nuevamente las magnitudes ajustadas
print("\nMagnitudes Ajustadas:")
mostrar_magnitudes(cesta_mundo)
```

```
# @title Texto de título predeterminado
import random
from fractions import Fraction
import math

# Definimos una clase para representar magnitudes
class Magnitud:
    def __init__(self, nombre, valor, tipo, cualidad):
        self.nombre = nombre
        self.valor = valor
        self.tipo = tipo  # Puede ser 'positiva' o 'negativa'
        self.cualidad = cualidad  # Puede ser 'cuantitativa' o 'cualitativa'

    def __str__(self):
        return f"{self.nombre} ({self.tipo}, {self.cualidad}): {self.valor} miles de millones de personas"

# Definimos la población total
poblacion_total = 8025  # miles de millones de personas

# Creamos una lista de magnitudes ajustadas a la población
cesta_mundo = [
    Magnitud("Salud", round(random.uniform(500, 1000), 2), 'positiva', 'cuantitativa'),
    Magnitud("Catástrofes", round(random.uniform(200, 500), 2), 'negativa', 'cuantitativa'),
    Magnitud("Pandemias", round(random.uniform(200, 500), 2), 'negativa', 'cuantitativa'),
    Magnitud("Accidentes", round(random.uniform(-100, -50), 2), 'negativa', 'cuantitativa'),
    Magnitud("Muertes de Hombres", round(random.uniform(1000, 1500), 2), 'positiva', 'cuantitativa'),
    Magnitud("Muertes de Mujeres", round(random.uniform(1500, 2000), 2), 'negativa', 'cuantitativa'),
    Magnitud("Distancia Recorrida a la Derecha", round(random.uniform(300, 700), 2), 'positiva', 'cualitativa'),
    Magnitud("Distancia Recorrida a la Izquierda", round(random.uniform(300, 700), 2), 'negativa', 'cualitativa'),
    Magnitud("Debe", round(random.uniform(50, 100), 2), 'negativa', 'cualitativa'),
    Magnitud("Haber", round(random.uniform(50, 100), 2), 'positiva', 'cualitativa'),
    Magnitud("Tiempo Pasado", round(random.uniform(1000, 2000), 2), 'negativa', 'cualitativa'),
    Magnitud("Tiempo Futuro", round(random.uniform(1000, 2000), 2), 'positiva', 'cualitativa')
]

# Función para mostrar las magnitudes
def mostrar_magnitudes(magnitudes):
    for magnitud in magnitudes:
        print(magnitud)

# Función para realizar la sustracción aritmética con manejo de casos especiales
def sustraccion_aritmetica(minuendo, sustraendo):
    resultado = minuendo.valor - sustraendo.valor
    if resultado < 0:
        parte_restante = -resultado
        print(f"Operación imposible: sustraendo {sustraendo.valor} es mayor que minuendo {minuendo.valor}.")
        print(f"Restamos la parte igual ({minuendo.valor}) y queda un número negativo ({-parte_restante}).")
        print(f"El número negativo {-parte_restante} cumple la relación fundamental porque {minuendo.valor} - {sustraendo.valor} = {resultado}.")
    else:
        print(f"Resultado de la resta: {resultado}")
    return resultado

# Ejemplo de uso de la sustracción aritmética
minuendo = Magnitud("Muertes de Hombres", 2500, 'positiva', 'cuantitativa')
sustraendo = Magnitud("Muertes de Mujeres", 3500, 'negativa', 'cuantitativa')
resultado_sustraccion = sustraccion_aritmetica(minuendo, sustraendo)
print(f"\nResultado de la Sustracción: {resultado_sustraccion}")

# Mostramos las magnitudes con valores continuos
print("\nMagnitudes con Valores Continuos:")
mostrar_magnitudes(cesta_mundo)

# Función para medir el conjunto continuo y obtener un número conmensurable
def medir_conjunto(magnitudes):
    total_valor = sum(magnitud.valor for magnitud in magnitudes)
    return round(total_valor, 2)

# Contamos unidades discretas y mostramos el resultado
def contar_unidades_discretas(magnitudes, unidad_objeto):
    conteo = sum(magnitud.valor // unidad_objeto for magnitud in magnitudes)
    return conteo

# Contamos unidades continuas y mostramos el resultado
def contar_unidades_continuas(magnitudes, unidad_medida):
    conteo = sum(magnitud.valor / unidad_medida for magnitud in magnitudes)
    return round(conteo, 2)

# Convertimos las mediciones a fracciones
def medir_fracciones(magnitudes, unidad_medida):
    fracciones = [Fraction(magnitud.valor).limit_denominator() for magnitud in magnitudes]
    total_fraccion = sum(fracciones)
    return total_fraccion / Fraction(unidad_medida)

# Medimos con números irracionales
def medir_irracionales(magnitudes, unidad_medida):
    irracionales = [math.sqrt(magnitud.valor) for magnitud in magnitudes]
    total_irracional = sum(irracionales)
    return round(total_irracional / unidad_medida, 2)

# Medimos con números racionales
def medir_racionales(magnitudes, unidad_medida):
    racionales = [Fraction(magnitud.valor).limit_denominator() for magnitud in magnitudes]
    total_racional = sum(racionales)
    return total_racional / Fraction(unidad_medida)

# Manejo de casos especiales
def manejar_casos_especiales(magnitudes, unidad_medida):
    for magnitud in magnitudes:
        if magnitud.valor % unidad_medida != 0:
            print(f"Advertencia: {magnitud.nombre} no contiene exactamente la unidad o sus partes.")
            magnitud.valor = round(magnitud.valor / unidad_medida) * unidad_medida
            print(f"Solución propuesta: Ajustar el valor a {magnitud.valor}.")

# Comparación de distancias recorridas para obtener un resultado positivo y negativo
distancia_derecha = Magnitud("Distancia Recorrida a la Derecha", 400, 'positiva', 'cualitativa')
distancia_izquierda = Magnitud("Distancia Recorrida a la Izquierda", 600, 'negativa', 'cualitativa')
resultado_comparacion_distancia = sustraccion_aritmetica(distancia_derecha, distancia_izquierda)
print(f"\nResultado de la Comparación de Distancias (Derecha - Izquierda): {resultado_comparacion_distancia}")

# Medimos el conjunto continuo y mostramos el resultado
resultado_medicion = medir_conjunto(cesta_mundo)
print(f"\nValor Conmensurable del Conjunto: {resultado_medicion} miles de millones de personas")

# Número de Unidades Discretas
unidad_objeto = 200  # Unidad arbitraria para las cantidades discretas
resultado_conteo_discreto = contar_unidades_discretas(cesta_mundo, unidad_objeto)
print(f"\nNúmero de Unidades Discretas: {resultado_conteo_discreto}")

# Número de Unidades Continuas
unidad_medida = 1.0  # Unidad arbitraria para las cantidades continuas
resultado_conteo_continuo = contar_unidades_continuas(cesta_mundo, unidad_medida)
print(f"Número de Unidades Continuas: {resultado_conteo_continuo}")

# Medición en Fracciones
resultado_fracciones = medir_fracciones(cesta_mundo, unidad_medida)
print(f"\nMedición en Fracciones: {resultado_fracciones}")

# Medición con Números Irracionales
resultado_irracionales = medir_irracionales(cesta_mundo, unidad_medida)
print(f"Medición con Números Irracionales: {resultado_irracionales}")

# Medición con Números Racionales
resultado_racionales = medir_racionales(cesta_mundo, unidad_medida)
print(f"\nMedición con Números Racionales: {resultado_racionales}")

# Manejo de casos especiales y solución
print("\nManejo de Casos Especiales:")
manejar_casos_especiales(cesta_mundo, unidad_medida)

# Mostramos nuevamente las magnitudes ajustadas
print("\nMagnitudes Ajustadas:")
mostrar_magnitudes(cesta_mundo)
```
